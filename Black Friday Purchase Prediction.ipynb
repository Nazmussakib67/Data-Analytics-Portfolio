### Cleaning, preparing and model training

# Problem Statement
A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month.
The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Import train data

df_train=pd.read_csv('train.csv/train.csv')
df_train.head()
```

```python
## Import test data

df_test=pd.read_csv('test.csv/test.csv')
df_test.head()
```

```python
## Merge both train and test data

#df=df_train.append(df_test)
#df.head()

## The error "DataFrame' object has no attribute 'append'" occurs because the append method was removed from Pandas version 1.4.0. This change was made to improve performance and avoid potential issues with index alignment.
```

```python
## Merge both train and test data (use concat instead append because append was removed from updated version)

df=pd.concat([df_train,df_test])
df.head()
```

```python
df.info()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
pd.get_dummies(df['Gender'])
```

```python
df['Gender']=df['Gender'].map({'F':0,'M':1})

#We can also use label encoder to change the data like this
```

```python
df.head()
```

```python
## Handle categorical feature age
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
##Fixing categorical city_category

df_city=pd.get_dummies(df['City_Category'],drop_first=True)
```

```python
df_city.head()
```

```python
df_city = df_city.astype(int) #use this to get numerical insted of True False
```

```python
df=pd.concat([df,df_city],axis=1)
```

```python
df.drop('City_Category',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Missing Values

df.isnull().sum()
```

```python
## Focusing on replacing missing values

df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
#Best way to deal with missing value is replace missing value with mode

df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
#Now replacing missing value for product category 3
df['Product_Category_3'].unique()
```

```python
df['Product_Category_3'].value_counts()
```

```python
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
df.info()
```

```python
# Convert object into integers

df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```

```python
##Feature Scaling 
df_test=df[df['Purchase'].isnull()]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     x, y, test_size=0.33, random_state=42)
```

```python
X_train.drop('Product_ID',axis=1,inplace=True)
X_test.drop('Product_ID',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(X_train, y_train)
```

```python
forest.score(X_test, y_test)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor

deci = DecisionTreeRegressor()
deci.fit(X_train, y_train)
```

```python
deci.score(X_test, y_test)
```
