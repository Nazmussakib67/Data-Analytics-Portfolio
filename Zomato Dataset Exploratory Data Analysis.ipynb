```python
df.isnull().sum()
```

```python
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country=pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
df_final=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
df_final.head()
```

```python
## To Check data type

df_final.dtypes
```

```python
df_final.columns
```

```python
country_names=df_final.Country.value_counts().index
country_names
```

```python
country_value=df_final.Country.value_counts()
country_value
```

```python
## Pie Chart- Top 3 countries

plt.pie(country_value[:3],labels=country_names[:3],autopct='%1.2f%%')

plt.show()
```

### Observation of pie chart- Zomato maximum transactions are from India, then USA then UK


```python
df_final.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings=df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```
